## <font color = 'Blue'>Question 1A:</font>

**(a) Create a list of 10 random integers between 0 and 100.**

```python
#Import 'random' library
import random

#Instantiate a list
my_random_list = []

#Loop to iterate over the range and 
#Append randomly generated numbers to the list
for i in range (10):
    my_random_list.append(random.randint(0, 100))

#Print the output
print(my_random_list)
print(type(my_random_list))
```

```python
#Another approach
#Instantiate a list
my_random_list1=[]

#Loop to iterate over the range using list comprehension
my_random_list1 = [random.randrange(0, 100, 1) for i in range(10)]

#Print the output
print(my_random_list1)
```

**(b) Create a numpy array of 10 random integers between 0 and 100.**

```python
#Import numpy library
import numpy as np

#Instantiate a list
random_list = []

#Loop to iterate over the range and 
#Append randomly generated numbers to the list and
#Convert list to numpy array
for i in range (10):
    random_list.append(random.randint(0, 100))
    my_random_array = np.array(random_list)

#Print the output
print(my_random_array)
print(type(my_random_array))
```

**(c) Calculate and print the sum of all the numbers in the array and the sum of all the numbers in the list.**

```python
#Method to add elements in a list or array
def _sum(array):
    
    sum = 0
    
    for i in array:
        sum = sum + i
    
    return(sum)
```

```python
#Print the output
print('Sum of list is: ', _sum(my_random_list))
print(my_random_list)
print('----------------------------------------------')
print('Sum of array is: ', _sum(my_random_array))
print(my_random_array)
```

## <font color = 'Blue'>Question 1B:</font>

**(a) Create a list, numpy array and a dictionary in python and initialize these to any set of values.**

**Iterate over the members of each of these objects and print them.**

```python
#Instantiate a list, array and a dictionary
my_list = [1, 5, 3, 10, 6, 38, 12, 45, 7, 18]
my_array = np.array([1, 3, 6, 12, 7, 5, 10, 38, 45, 18])
my_dict = {1: 'Mercedes', 2: 'Hyundai', 3: 'Honda', 4: 'Nissan', 5: 'Kia'}

#Loop to iterate over the elements in the list
print('Iterate over elements in a list:')
for i in my_list:
    print(i)
print('--------------')
    
#Loop to iterate over the elements in the array
print('Iterate over elements in an array:')
for j in my_array:
    print(j)
print('--------------')

#Loop to iterate over the elements in the dictionary
print('Iterate over elements in a dictionary:')
for k in my_dict:
    print(k, ':', my_dict[k])
```

**(b) Create a two-dimensional array in python. Iterate over the array to print out each row, each column and each value.**

```python
#Instantiate a 2D array and fill with random numbers
two_dim_array = np.random.randint(1, 40, size=(4, 4))
two_dim_array
```

```python
#Show the shape of the array
print('Shape of the 2D array : ', two_dim_array.shape)
```

```python
print('Iterate over rows output:')
for rows in two_dim_array:
    print(rows)
print('---------------------------------------')
print('Iterate over columns output:')
for cols in two_dim_array.T:
    print(cols)
print('---------------------------------------')
print('Iterate over each element in the 2D array output:')
for a in two_dim_array:
    for b in a:
        print(b)
```

```python
#Read 'housing' csv file
housing = pd.read_csv('housing.csv')
```

**1.Modify the code so that the housing dataframe only include rows where the median_house_value is less than 500,000 and the housing_median_age is less than 50.**

```python
#Filter the dataframe using loc and the given conditions
housing = housing.loc[(housing['median_house_value'] < 500000) & (housing['housing_median_age'] < 50)]
housing
```

```python
housing[housing.housing_median_age == 25]
```

**2.Now plot all the histograms (as we did in that notebook), and you will see that the “spikes” at the right for those two variables have disappeared.**

```python
#Import matplotlib library
import matplotlib
import matplotlib.pyplot as plt

#Note 'bins = 50'
housing.hist(bins=50,figsize=(20,15))
plt.show()
```

**3.Look at the histogram plot for “housing_median_age” now. You will see a small “gap” in the middle. Why do you think the gap is there? Modify something in the code and re-plot the histograms so that this “gap” is removed.**

```python
#Showing the boundry and frequencies for 23rd, 24th, 25th and 26th bins
data_bin = housing.housing_median_age

counts, edges, plot = plt.hist(data_bin, bins=50)
print('\n23rd bin frequency:', counts[22], '\n24th bin frequency:', counts[23], '\n25th bin frequency:', counts[24], '\n26th bin frequency:', counts[25])
print('---------------------------')
limit1 = edges[23] - 0.04
limit2 = edges[24] - 0.04
limit3 = edges[25] - 0.04
limit4 = edges[26] - 0.04

print('24th bin boundry', limit1, '\n25th bin boundry', limit2, '\n26th bin boundry', limit3, '\n27th bin boundry', limit4)

#Showing frequecy of the missing value
print('---------------------------')
val25 = pd.DataFrame(housing.housing_median_age.value_counts()).reset_index()
val25.iloc[8]
```

**Explanation for the gap in the histogram**

```python
#Changed 'bin = 49' and the gap is removed
housing.hist(bins=49, figsize=(20,15))
plt.show()

#Binning in histogram is where the data is grouped into "bins" of equal width.
#The initial bin width was 50, but after applying the filters from the previous question, the values we had were 49.

#From the plot above, the gap occured as the bin width was 50 but the actual values were 49 (Dividing value range 1-49 by 50 equals single bin width of 0.96).
#This causes a value (25) to fall within the boundries of bin on the right (Within 26th bin) which causes the gap. Changing the bin width to 49 fixed the issue.
```
